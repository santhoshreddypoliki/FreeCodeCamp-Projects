```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sts
import numpy as np
```

```python
df = pd.read_csv("c://users/santhosh reddy/documents/epa_sea_level.csv")
```

```python
df.head()
```

```python
df = df.rename(columns={'Year':'year', 'CSIRO Adjusted Sea Level':'csiro_adjusted_sea_level', 'Lower Error Bound':'lower_error_bound', 'Upper Error Bound':'upper_error_bound', 'NOAA Adjusted Sea Level':'noaa_adjusted_sea_level'})
```

```python
plt.figure(figsize=(15,10))
sns.scatterplot(x = df.year, y = df.csiro_adjusted_sea_level)
```

```python
line = sts.linregress(x = df.year , y = df.csiro_adjusted_sea_level)
```

```python
xA = np.arange(df['year'].min(),2050,1)
yA = xA*line.slope + line.intercept
```

```python
plt.plot(xA , yA)
sns.scatterplot(x = df.year, y = df.csiro_adjusted_sea_level)
```

```python
df_2000 = df[df['year']>=2000]
```

```python
lineb = sts.linregress(x = df_2000.year , y = df_2000.csiro_adjusted_sea_level)
```

```python
xb = np.arange(2000,2050,1)
yb = xb*lineb.slope + lineb.intercept
```

```python
plt.plot(xb , yb)
sns.scatterplot(x = df.year, y = df.csiro_adjusted_sea_level)
```

```python
plt.plot(xA , yA)
plt.plot(xb , yb)
sns.scatterplot(x = df.year, y = df.csiro_adjusted_sea_level)
```
